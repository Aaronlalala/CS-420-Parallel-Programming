```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df1 = pd.read_excel("results.xlsx", sheet_name="strong")
df2 = pd.read_excel("results.xlsx", sheet_name="weak")
df3 = pd.read_excel("results.xlsx", sheet_name="multiple nodes")
```

```python
df1
```

```python
plt.figure(dpi=100)
plt.plot(df1["Number of Processes"], df1["Execution Time"])
plt.title("Strong Scaling Single Node")
plt.xticks([1, 2, 4, 8, 16])
plt.xlabel("Number of Processes")
plt.ylabel("Execution Time")
plt.grid()
plt.show()
```

```python
df2.drop(index=1)
```

```python
plt.figure(dpi=100)
plt.plot(df2["Number of Processes"], df2["Execution Time"])
plt.plot(df2.drop(index=1)["Number of Processes"], df2.drop(index=1)["Execution Time"]+0.2)
plt.title("Weak Scaling")
plt.xticks([2, 8, 32])
plt.xlabel("Number of Processes")
plt.ylabel("Execution Time")
plt.legend(["actual", "ideal"])
plt.grid()
plt.show()
```

```python
plt.figure(dpi=100)
plt.plot(df3["Number of Processes"], df3["Execution Time"])
plt.title("Strong Scaling Multiple Nodes")
plt.xticks([1, 2, 4, 8])
plt.xlabel("Number of Processes")
plt.ylabel("Execution Time")
plt.grid()
plt.show()
```
