```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_excel("data.xlsx", sheet_name="Q1")
df2 = pd.read_excel("data.xlsx", sheet_name="Q2")
df3 = pd.read_excel("data.xlsx", sheet_name="Q3")
```

```python
plt.figure(dpi=100)
plt.plot(df1["Number of Processes"], df1["Execution Time"])
plt.title("Strong Scaling Single Node")
plt.xlabel("Number of Processes")
plt.ylabel("Execution Time")
plt.show()
```

```python
plt.figure(dpi=100)
plt.plot(df2["Number of Processes"], df2["Execution Time"])
plt.title("Weak Scaling")
plt.xticks([4, 16, 64])
plt.xlabel("Number of Processes")
plt.ylabel("Execution Time")
plt.show()
```

```python
plt.figure(dpi=100)
plt.plot(df3["Number of Processes"], df3["Execution Time"])
plt.title("Strong Scaling Multiple Nodes")
plt.xticks([16, 32, 64, 128])
plt.xlabel("Number of Processes")
plt.ylabel("Execution Time")
plt.xscale('log')
plt.show()
```
